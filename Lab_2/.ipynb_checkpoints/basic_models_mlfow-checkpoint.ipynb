```python
import numpy as np
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt
import mlflow
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


import warnings
warnings.filterwarnings("ignore")
```

# DATA + initialize experiments


```python
df = pd.read_csv('data/preprocessed_data.csv',parse_dates=['ds'])[['ds','y']]
df.head()
```

```python
train, test = df[0:-365], df[-365:]
```

```python
#see the TS plot

plt.rc('font', size=12)
fig, ax = plt.subplots(figsize=(20, 6))

ax.plot(train.ds, train.y, color='orange', label='train')
ax.plot(test.ds, test.y, color='green', label='test')

ax.set_xlabel('Date')
ax.set_ylabel('Mean Temp in Madrid')
ax.legend(loc='upper left')
```

```python
mlflow.set_tracking_uri('sqlite:///mlflow.db')
```

```python
mlflow.set_experiment('timeseries-experiment')
```

# Prophet project

```python
fourier_order = [1,3,5,7,10,12,15,17]
```

```python
for i in fourier_order:
    with mlflow.start_run():
        mlflow.set_tags({"Model":"Prophet", "Train Data": "3 years"})

        mlflow.log_params({'fourier_order':i})

        model = Prophet() #default include weekly and yearly seasonalities

        #Fit with default settings
        model.add_seasonality(name='yearly', period=365, fourier_order=i)
        model.fit(train)

        future = model.make_future_dataframe(periods=365) #freq='D'
        forecast = model.predict(future)
        
        y_pred = forecast['yhat'][-365:]
        y_test = test.y
        
        # Mean Absolute Error (MAE)
        mae = np.round(mean_absolute_error(y_test, y_pred),4)
        mlflow.log_metric('MAE', mae)
        
        # Mean Squared Error (MSE)
        mse = np.round(mean_squared_error(y_test, y_pred),4)
        mlflow.log_metric('MSE', mse)

        # Root Mean Squared Error (RMSE)
        rmse = np.round(np.sqrt(mse),4)
        mlflow.log_metric('RMSE', rmse)

        # R² Score (Coefficient of Determination)
        r2 = np.round(r2_score(y_test, y_pred),4)
        mlflow.log_metric('R2', r2)

    mlflow.end_run()
```

# ETS

```python
import statsmodels as sm
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
data = pd.read_csv('data/preprocessed_data.csv',parse_dates=['ds'], index_col='ds')['y']
train, test = data[0:-365], data[-365:]
```

```python
trend = ['add', 'mul', None]
season = ['add', 'mul', None]
damp_trend = [False, True]
```

```python
for t in trend:
    for s in season:
        for d in damp_trend:
            with mlflow.start_run():
                mlflow.set_tags({"Model":"ETS", 
                                 "Trend":t,
                                 "Seasonality":s,
                                 "Damped Trend":d,
                                 "Train Data": "3 years"})

                if t is None:
                    model = ExponentialSmoothing(train, trend=t, seasonal=s, seasonal_periods=365, freq='D').fit()
                else:
                    model = ExponentialSmoothing(train, trend=t, seasonal=s, seasonal_periods=365, damped_trend=d, freq='D').fit()
                    
                forecast = pd.DataFrame(model.forecast(len(test)))


                y_pred = forecast
                y_test = test

                # Mean Absolute Error (MAE)
                mae = np.round(mean_absolute_error(y_test, y_pred),4)
                mlflow.log_metric('MAE', mae)

                # Mean Squared Error (MSE)
                mse = np.round(mean_squared_error(y_test, y_pred),4)
                mlflow.log_metric('MSE', mse)

                # Root Mean Squared Error (RMSE)
                rmse = np.round(np.sqrt(mse),4)
                mlflow.log_metric('RMSE', rmse)

                # R² Score (Coefficient of Determination)
                r2 = np.round(r2_score(y_test, y_pred),4)
                mlflow.log_metric('R2', r2)

            mlflow.end_run()
```

```python
plt.plot(train, label='history', color='grey')
plt.plot(test, label='test')
plt.plot(y_pred, label='additive ETS')

plt.legend(loc='upper left', fontsize=8)
plt.show()
```

# ARIMA

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
p_values = [0, 1, 2, 3, 4, 5]
d_values = [0, 1, 2]
q_values = [0, 1, 2, 3, 4, 5]
```

```python

for p in p_values:
    for d in d_values:
        for q in q_values:
            order = (p, d, q)
            try:
                with mlflow.start_run():
                    mlflow.set_tags({"Model":"ARIMA", "Train Data": "3 years"})
                    mlflow.log_params({'p':p})
                    mlflow.log_params({'q':q})
                    mlflow.log_params({'d':d})
                    model = ARIMA(train, order=order,freq='D')
                    model_fit = model.fit()
                    # Predict (in-sample prediction)
                    y_pred = model_fit.forecast(365)
                    y_test = test

                    # Mean Absolute Error (MAE)
                    mae = np.round(mean_absolute_error(y_test, y_pred),4)
                    mlflow.log_metric('MAE', mae)

                    # Mean Squared Error (MSE)
                    mse = np.round(mean_squared_error(y_test, y_pred),4)
                    mlflow.log_metric('MSE', mse)

                    # Root Mean Squared Error (RMSE)
                    rmse = np.round(np.sqrt(mse),4)
                    mlflow.log_metric('RMSE', rmse)

                    # R² Score (Coefficient of Determination)
                    r2 = np.round(r2_score(y_test, y_pred),4)
                    mlflow.log_metric('R2', r2)
                
                mlflow.end_run()
            except:
                continue
            


```
